```python
joint_df = pd.DataFrame({
    'ngram': common_words,
    'frequency': [freq_dict_b_w[f'{w}']/len(freq_dict_b_w) for w in freq_dict_b_w.keys() if w in common_words],
    'freq_diff': [(freq_dict_b_w[f'{w}']/len(freq_dict_b_w))-(freq_dict_p_k[f'{w}']/len(freq_dict_p_k)) for w in freq_dict_b_w.keys() if w in common_words]
})
```
